```python
import pandas as pd
import numpy as np

# Load data
data = pd.read_csv("heart_disease_uci.csv")

# Drop unnecessary columns
data = data.drop(columns=['id', 'dataset'])

# Convert categorical to proper formats
data['sex'] = data['sex'].map({'Male': 1, 'Female': 0})
data['fbs'] = data['fbs'].map({'True': 1, 'False': 0})
data['exang'] = data['exang'].map({'True': 1, 'False': 0})

# Fill numeric missing values with median
num_cols = ['trestbps', 'chol', 'thalch', 'oldpeak', 'ca']
for col in num_cols:
    data[col] = data[col].fillna(data[col].median())

# Fill categorical missing values with mode (most common value)
cat_cols = ['cp', 'restecg', 'slope', 'thal']
for col in cat_cols:
    data[col] = data[col].fillna(data[col].mode()[0])

# Convert 'num' to binary: 0 = no disease, 1 = disease present
data['num'] = data['num'].apply(lambda x: 1 if x > 0 else 0)

# One-hot encode categorical columns
data = pd.get_dummies(data, columns=['cp', 'restecg', 'slope', 'thal'], drop_first=True)

# Check again
print(data.info())
print("\n✅ Missing values after cleaning:\n", data.isnull().sum())
# Convert True/False or Yes/No to 1/0
data['fbs'] = data['fbs'].replace({'True': 1, 'False': 0})
data['exang'] = data['exang'].replace({'True': 1, 'False': 0})

# If they are still all NaN, fill with 0 (safe assumption)
data['fbs'] = data['fbs'].fillna(0)
data['exang'] = data['exang'].fillna(0)
data.isnull().sum()


```

```python

#selecting dependent and independent varaiable
x= data.drop(columns=['num'])
y = data['num']
x = x.fillna(x.mean())   # fill missing numeric values with mean
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print("Training data shape:", x_train.shape)
print("Testing data shape:", x_test.shape)
print("Training data shape:", y_train.shape)
print("Testing data shape:", y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score

# Predictions
y_pred = model.predict(x_test)

# Evaluation
print("✅ Accuracy:", accuracy_score(y_test, y_pred))
print("\n📊 Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\n📋 Classification Report:\n", classification_report(y_test, y_pred))

# ROC-AUC
y_proba = model.predict_proba(x_test)[:, 1]
print("\n🎯 ROC-AUC Score:", roc_auc_score(y_test, y_proba))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
```

```python
models = {
    "Logistic Regression": LogisticRegression(),
    "Random Forest": RandomForestClassifier(),
    "SVM": SVC()
}

for name, model in models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    print(f"{name} Accuracy: {model.score(x_test, y_test):.2f}")
```

```python
age = float(input("Enter age (29-77): "))
sex = int(input("Enter sex (Male=1, Female=0): "))
trestbps = float(input("Resting blood pressure (trestbps, 90-200): "))
chol = float(input("Cholesterol (chol, 150-350): "))
thalch = float(input("Max heart rate (thalch, 70-210): "))
oldpeak = float(input("ST depression (oldpeak, 0.0-6.0): "))

# Binary inputs
fbs = int(input("Fasting blood sugar >120 mg/dl? (Yes=1, No=0): "))
exang = int(input("Exercise induced angina? (Yes=1, No=0): "))

# Categorical inputs
cp = input("Chest pain type (typical angina / atypical angina / non-anginal): ")
restecg = input("Resting ECG (normal / st-t abnormality / lv hypertrophy): ")
slope = input("Slope of peak exercise ST segment (upsloping / flat / downsloping): ")
thal = input("Thalassemia (normal / fixed defect / reversable defect): ")

# 9️⃣ Prepare input array
input_array = np.zeros(len(x.columns))

# Map numeric features
cols_numeric = ['age','sex','trestbps','chol','thalch','oldpeak','fbs','exang']
values_numeric = [age, sex, trestbps, chol, thalch, oldpeak, fbs, exang]
for col, val in zip(cols_numeric, values_numeric):
    input_array[x.columns.get_loc(col)] = val

# Map one-hot categorical features
if 'cp_'+cp in x.columns:
    input_array[x.columns.get_loc('cp_'+cp)] = 1
if 'restecg_'+restecg in x.columns:
    input_array[x.columns.get_loc('restecg_'+restecg)] = 1
if 'slope_'+slope in x.columns:
    input_array[x.columns.get_loc('slope_'+slope)] = 1
if 'thal_'+thal in x.columns:
    input_array[x.columns.get_loc('thal_'+thal)] = 1

# Convert to DataFrame to match scaler's expected input
input_df = pd.DataFrame([input_array], columns=x.columns)

# 10️⃣ Scale user input
input_scaled = scaler.transform(input_df)

# 11️⃣ Predict
prediction = model.predict(input_scaled)[0]
probability = model.predict_proba(input_scaled)[0][1]

# 12️⃣ Show result
if prediction == 1:
    print(f"⚠️ Risk of heart disease (Probability: {probability:.2f})")
else:
    print(f"✅ Not likely to have heart disease (Probability: {probability:.2f})")
```
